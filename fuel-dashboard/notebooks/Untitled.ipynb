```python
import os
import pandas as pd


BASE_DATA_PATH = "../../data/"

dfs = []
for parquet in os.listdir(BASE_DATA_PATH):
    parquet_path = os.path.join(BASE_DATA_PATH, parquet)
    dfs.append(pd.read_parquet(parquet_path))

df = pd.concat(dfs).sort_values("timestamp", ascending=False)
```
